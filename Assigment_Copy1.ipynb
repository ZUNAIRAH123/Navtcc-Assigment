```python
from sklearn.model_selection import train_test_split
```

```python
X= df.drop('np6024', axis = 1)
Y = df['np6024']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_TRAIN,X_TEST, Y_TRAIN,Y_TEST = train_test_split(X, Y, test_size=0.25, random_state=30)
```

```python
print("Size of Train X", len(X_TRAIN))
print("Size of Train y", len(Y_TRAIN))
print("Size of Test y", len(X_TEST))
print("Size of Test Y", len(Y_TEST))
```

# GaussianNB Base Classifier

```python
from sklearn.naive_bayes import GaussianNB
eman=GaussianNB()

# Discretize Y_TRAIN and Y_TEST into 5 bins
Y_TRAIN_binned = pd.cut(Y_TRAIN, bins=5, labels=False, include_lowest=True)
Y_TEST_binned = pd.cut(Y_TEST, bins=5, labels=False, include_lowest=True)

eman.fit(X_TRAIN, Y_TRAIN_binned)
```

```python
# Predict on training data
training_data_prediction = eman.predict(X_TRAIN)
# Print predictions
print(training_data_prediction)
```

```python
from sklearn.metrics import accuracy_score
training_accuracy = accuracy_score(Y_TRAIN_binned, training_data_prediction)
print("Training Accuracy:", training_accuracy)
```

# Prediction on Train Data

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_TRAIN_binned, training_data_prediction)
print("The Accuracy of Prediction is ", accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix(Y_TRAIN_binned, training_data_prediction)
```

```python
import matplotlib.pyplot as plt
cm = confusion_matrix(Y_TRAIN_binned, training_data_prediction)
fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(cm)
ax.xaxis.set(ticks=range(len(eman.classes_)), ticklabels=[f'Predicted {c}' for c in eman.classes_])
ax.yaxis.set(ticks=range(len(eman.classes_)), ticklabels=[f'Actual {c}' for c in eman.classes_])
for i in range(len(eman.classes_)):
    for j in range(len(eman.classes_)):
        ax.text(j, i, cm[i, j], ha='center', va='center',
                color='white')
plt.show()
```

```python
report = classification_report(Y_TRAIN_binned, training_data_prediction)
print(report)
```
